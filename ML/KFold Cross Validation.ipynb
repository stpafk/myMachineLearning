**KFOLD WITH DATASETS**

```python
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.svm import SVC
from sklearn.ensemble import RandomForestClassifier
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeClassifier
import numpy as np
from sklearn.datasets import load_digits

digits = load_digits() # classify digits dataset and use different models and evaluate each performance
```

```python
X_train, X_test, y_train, y_test = train_test_split(digits.data, digits.target, test_size=0.3)
```

```python
lr = LogisticRegression(solver='lbfgs', max_iter=10000)
lr.fit(X_train, y_train)
lr.score(X_test, y_test)
```

```python
svm = SVC()
svm.fit(X_train, y_train)
svm.score(X_test, y_test)
```

```python
rf = RandomForestClassifier()
rf.fit(X_train, y_train)
rf.score(X_test, y_test)
```

```python
#That's the problem with train test split, you run multiple times and get different results
```

```python
from sklearn.model_selection import KFold
kf = KFold(n_splits=3) # fold which separate and test data
kf
```

```python
for train_index, test_index in kf.split([1,2,3,4,5,6,7,8,9]):
    print(train_index, test_index)
```

```python
#this is how it tests: the test_index is removed from the array, and it is used as
#a training model. for all the iteration, it does the same thing
```

```python
def get_score(model, X_train, X_test, y_train, y_test):
    model.fit(X_train, y_train)
    return model.score(X_test, y_test)
```

```python
get_score(SVC(), X_train, X_test, y_train, y_test)
```

```python
from sklearn.model_selection import StratifiedKFold
fold = StratifiedKFold(n_splits=3)
```

```python
scores_l = []
scores_svm = []
scores_rf = []

for train_index, test_index in kf.split(digits.data):
    X_train, X_test, y_train, y_test = digits.data[train_index], digits.data[test_index], \
                                        digits.target[train_index], digits.target[test_index]
    scores_l.append(get_score(LogisticRegression(max_iter=100000), X_train, X_test, y_train, y_test))
    scores_svm.append(get_score(SVC(), X_train, X_test, y_train, y_test))
    scores_rf.append(get_score(RandomForestClassifier(n_estimators=40), X_train, X_test, y_train, y_test))
```

```python
scores_svm
```

```python
scores_rf
```

```python
from sklearn.model_selection import cross_val_score #does the same thing as iter
```

```python
cross_val_score(SVC(), digits.data, digits.target)
```

```python
from sklearn.datasets import load_iris
iris = load_iris()
```

```python
lr = cross_val_score(LogisticRegression(max_iter=100000), iris.data, iris.target)
np.average(lr)
```

```python
svm = cross_val_score(SVC(), iris.data, iris.target)
np.average(svm)
```

```python
dt = cross_val_score(DecisionTreeClassifier(), iris.data, iris.target)
np.average(dt)
```

```python
rf = cross_val_score(RandomForestClassifier(n_estimators=40), iris.data, iris.target)
np.average(rf)
```

```python
from sklearn.linear_model import LinearRegression
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```
