titanic

```python
import pandas as pd
```

```python
t = pd.read_csv('titanic.csv')
t.head(6)
```

```python
x = t.drop(['PassengerId', 'Name', 'SibSp', 'Parch', 'Ticket', 'Cabin', 'Embarked'], axis='columns')
```

```python
x = x.drop(['Survived'], axis='columns')
y = t['Survived']
```

```python
dummies = pd.get_dummies(x.Sex)
```

```python
x = pd.concat([x, dummies], axis='columns')
```

```python
x.drop(['Sex'], axis='columns', inplace=True)
```

```python
x.Age = x.Age.fillna(x.Age.mean())
```

```python
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split(x, y, test_size=0.3)
```

```python
print(len(X_test))
print(len(X_train))
```

```python
from sklearn.naive_bayes import GaussianNB
model = GaussianNB()
```

```python
model.fit(X_train, y_train)
```

```python
model.score(X_test, y_test)
```

```python
y_test[:10]
```

```python
model.predict(X_test[:10])
```

```python
model.predict_proba(X_test[:10])
```

email spam

```python
s = pd.read_csv('spam.csv')
s.head(5)
```

```python
s.groupby('Category').describe()
```

```python
s['spam'] = s['Category'].apply(lambda x: 1 if x=='spam' else 0)
```

```python
s.head()
```

```python
X_train, X_test, y_train, y_test = train_test_split(s.Message, s.spam, test_size=0.3)
```

```python
#countvectorizer technic
#bernoulli naive bayes == when features are binary numbers
#multinomial naive bayes == when the data is discret(statistical), meaning it has a range and frequency (ex 1to5)
#gaussian naive bayes == when data has a normal distribution 
```

```python
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.naive_bayes import MultinomialNB
```

```python
v = CountVectorizer()
model = MultinomialNB()

#countvectorizer
X_train_count = v.fit_transform(X_train.values)
X_train_count.toarray()[:3]

#multinomialnb
model.fit(X_train_count, y_train)
```

```python
X_test_count = v.transform(X_test)
model.score(X_test_count, y_test)
```

```python
#simplifying the code using the API
from sklearn.pipeline import Pipeline
clf = Pipeline([
    ('vectorizer', CountVectorizer()),
    ('nb', MultinomialNB())
])
```

```python
clf.fit(X_train, y_train)
```

```python
clf.score(X_test, y_test)
```
