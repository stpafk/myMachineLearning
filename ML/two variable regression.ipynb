```python
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np
from sklearn import linear_model
```

```python
h = pd.read_csv('homeprices.csv')
h
```

```python
import math

bed_med = math.floor(h.bedrooms.median())
bed_med
#h.drop([5], axis=0, inplace=True)
h
```

```python
h.bedrooms = h.bedrooms.fillna(bed_med)
```

```python
reg = linear_model.LinearRegression()
reg.fit(h[['area', 'bedrooms', 'age']], h.price)
reg.coef_ #
```

```python
reg.predict([[3000, 3, 15]]) 
```
